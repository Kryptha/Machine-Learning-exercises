```python
# By @kryptha

import pandas as pd
import matplotlib.pyplot as pl

df = pd.read_csv('ds_salaries.csv')
# Data frame cleaned of duplicated data and reset indexes
df = df[~df.duplicated(keep='first')]
df.reset_index(drop=True)
df.head(10)
```

```python
# Get general information of all data frame
df.describe()
```

```python
import seaborn as sns
sns.set()
pl.figure(figsize=(19,10))

histogram_jobs = sns.countplot(x='job_title', data=df, order = df['job_title'].value_counts().index)
histogram_jobs.set_xticklabels(histogram_jobs.get_xticklabels(),rotation=45, fontsize =10, ha='right')

# Conclusion: There is more people with Data Science, Data Engineer, Data Analyst and Machine Learning Engineer than other job title. 
# Exists more people difference beetwen top 4 job title and others.
```

```python

#SE : Senior-level
#MI : Mid-level
#EN : Entry-level
#EX : Executive-level

sns.countplot(x='experience_level', data=df, order = df['experience_level'].value_counts().index)
# Conclusion: There are less people with executive level because it takes more years to be an executive level
```

```python
sns.jointplot(x="experience_level", y="salary_in_usd", data=df)
# Conclusion: Doesn't exists a big difference beetwen experience level of people and their salaries. 
# Executive level of people include an atypical values that can affect in the average of them. In general, 
# there is a concentration in the range of 150,000 USD and 200,000 USD independent their experience level
```

```python
pl.figure(figsize=(20, 20))
sns.swarmplot(x="employment_type", y="salary_in_usd", data=df)
# Conclusion: In the data set, a big concentration of full employment type is considered
```

```python
pl.figure(figsize=(10,10))
sns.histplot(data=df['salary_in_usd'],kde=True)

#Conclusion: There is a greater number of people earning close to 100,000 USD annually.
```

```python
pl.figure(figsize=(20,10))
sns.histplot(data=df, x='salary_in_usd', hue='experience_level')

#Conclusion: A stric relationship doesn't exists beetwen a more expierienced level and salary 
```

```python
pl.figure(figsize=(20,10))
sns.histplot(data=df, x='salary_in_usd', hue='experience_level', multiple="dodge", shrink=.8)
```

```python
pl.figure(figsize=(20,10))
sns.histplot(data=df, x='salary_in_usd', hue='employment_type')
```

```python
pl.figure(figsize=(30,10))
df['job_count'] = df.groupby('job_title')['job_title'].transform('count')
df2 = df.groupby(['job_title', 'job_count'])['salary_in_usd'].mean().reset_index().sort_values(['salary_in_usd'], ascending=False)

ax = sns.scatterplot(x='job_title', y='salary_in_usd', data=df2)
ax.set_xlabel('Mean of salary(US)')
pl.margins(x=0)
pl.xticks(rotation=90)

# Scatter graphic with mean of salary in USD and job title 
```

```python
pl.figure(figsize=(30,10))
ax=sns.histplot(x='salary_in_usd', y='job_count',hue='job_title', data=df2, multiple='stack')
sns.move_legend(ax, "upper left", bbox_to_anchor=(.99, .99), title='Job Titles')
ax.set_xlabel('Mean of salary(US)')
sns.despine()

# Histogram graphic with mean salary in USD and people count by job title
# Conclusion: You would think in the graphic above there is a relationship beetwen job title and mean of salaries. But this
# graph shows that the highest means are in jobs with fewer people, while the most common job title still in tange between 100,000 USD
# and 200,000 USD
```
