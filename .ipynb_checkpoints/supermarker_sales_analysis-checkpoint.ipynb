```python
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sns
import numpy as np
sns.set()

df = pd.read_csv('Stores.csv')
# Clean data frame of duplicated data and reset their index 
df = df[~df.duplicated(keep='first')]
df.reset_index(drop=True)
df.head(10)
```

```python
df2 = df[['Store_Sales', 'Daily_Customer_Count', 'Items_Available']]
pl.figure(figsize=(20,10))
sns.histplot(data=df2, x='Store_Sales', y='Daily_Customer_Count')
```

```python
pl.figure(figsize=(20,10))
sns.histplot(data=df2, x='Store_Sales', y='Items_Available')
```

```python
bins =  np.arange(900,3000,100)
groups1 = df.groupby(pd.cut(df['Items_Available'],bins)).mean()
groups1['Store_Sales'].plot.line()
```

```python
bins =  np.arange(10,1560,100)
groups2 = df.groupby(pd.cut(df['Daily_Customer_Count'],bins)).mean()
groups2['Store_Sales'].plot.line()
```

```python
bins =  np.arange(0,3000,1000)
groups2 = df.groupby(pd.cut(df['Store_Area'],bins)).mean()
groups2['Store_Sales'].plot.line()
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

x = df[['Daily_Customer_Count', 'Items_Available', 'Store_Area']]
y = df['Store_Sales']

x[['Daily_Customer_Count', 'Items_Available', 'Store_Area']] = scale.fit_transform(x[['Daily_Customer_Count', 'Items_Available', 'Store_Area']].values)
x = sm.add_constant(x)
est = sm.OLS(y, x).fit()
print(est.summary())

# Conclusion: Exists a relationship beetwen items avalaible and store sales, but it's dependent of atypical values. 
# It's also considered a increasing relationship beetwen store area and store sales.
# Also the daily customer count affects the model but in less quantity than items availables.
```

```python
scaled = scale.transform([[1500, 200, 1]])
scaled = np.insert(scaled[0], 0, 1) # Need a new column of constant by pre-requisite for sklearn librery
print(f'Estimated sale of {int(est.predict(scaled)[0])} USD')
```
